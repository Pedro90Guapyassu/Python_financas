#Trabalhando com mais de uma ação 

```python
acoes = ['GOLL4.SA', 'CVCB3.SA', 'WEGE3.SA', 'MGLU3.SA', 'TOTS3.SA', 'BOVA11.SA']
acoes_df = pd.DataFrame()
for acao in acoes:
  acoes_df[acao] = data.DataReader(acao, data_source='yahoo', start='2015-01-01')['Close']
```

```python
acoes_df.head()
```

```python
acoes_df.isnull().sum()
```

```python
acoes_df = acoes_df.rename(columns={'GOLL4.SA':'GOLL','CVCB3.SA':'CVC','WEGE3.SA':'WEGE','MGLU3.SA':'MGLU','TOTS3.SA':'TOTS','BOVA11.SA':'BOVA'})
```

```python
acoes_df.head()
```

```python
acoes_df.to_csv('acoes.csv')
```

#Vizualizações

```python
sns.histplot(acoes_df['GOLL']); # o eixo x aparece como o preço das açoes e o eixo y mostra quantas vezes ela bate aquele preço
#voce também pode utilizar o parametro bins para indicar o numero de barras/divisoes(vide o proximo exemploc)
```

```python
sns.histplot(acoes_df['GOLL'], bins = 10);
```

```python
plt.figure(figsize= (10, 50))
i = 1
for i in np.arange(1, len(acoes_df.columns)):
  plt.subplot(7, 1, i + 1) #7 número de linhas(que será de 6)e 1 número de colunas
  sns.histplot(acoes_df[acoes_df.columns[i]], kde = True)
  plt.title(acoes_df.columns[i])
```

```python
plt.figure(figsize= (10, 50))
i = 1
for i in np.arange(1, len(acoes_df.columns)):
  plt.subplot(7, 1, i + 1) #7 número de linhas(que será de 6)e 1 número de colunas
  sns.boxplot(x =acoes_df[acoes_df.columns[i]])
  plt.title(acoes_df.columns[i])
```

```python
acoes_df.plot(x = 'Date', figsize= (15,7), title = 'Histórico do preço das açoes');
```

#Normalização


O normalizar o gráfico torna a comparação mais justa na hora de identificar qual das ações possibilitou um ganho maior

```python
#divisão de cada valor pelo valor inicial
acoes_df_normalizado = acoes_df.copy()
for i in acoes_df_normalizado.columns[1:]:
  acoes_df_normalizado[i] = acoes_df_normalizado[i] /acoes_df_normalizado[i][0]
```

```python
acoes_df_normalizado
```

```python
acoes_df_normalizado.plot(x = 'Date', figsize= (15,7), title = 'Histórico do preço das açoes')
```

```python
figura = px.line(title='Histórico preço das ações')
for i in acoes_df.columns[1:]:
  figura.add_scatter(x = acoes_df['Date'], y=acoes_df[i], name = i)
figura.show()
```

```python
figura = px.line(title='Histórico preço das ações')
for i in acoes_df_normalizado.columns[1:]:
  figura.add_scatter(x = acoes_df_normalizado['Date'], y = acoes_df_normalizado[i], name = i)
figura.show()

```
